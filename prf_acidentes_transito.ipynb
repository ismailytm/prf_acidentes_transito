# 1\. Introdução
Visualização e Predição de Severidade dos acidentes registrados pela Polícia Rodiviária Federal entre os anos de 2017 e 2022.

# 2\. Bibliotecas e Datasets

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
import pandas as pd
import glob

# Path of directory containing the CSV files
path = 'data'

# Get all CSV files in the directory
all_files = glob.glob(path + "/*.csv")

# Read all CSV files and store them in a list
data_list = []
for filename in all_files:
    df = pd.read_csv(filename, sep=';', encoding='cp1252')
    data_list.append(df)

# Concatenate all CSV files into a single DataFrame
merged_df = pd.concat(data_list, axis=0, ignore_index=True)

# Save as a newfile
merged_df.to_csv('data\\merged_acidentes2017a2022.csv')

# Display the merged DataFrame
merged_df.head()
```
